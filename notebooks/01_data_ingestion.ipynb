```python
# import packages
import pandas as pd
from pathlib import Path

# setup directory to load data
RAW_DIR = Path("../data/raw")
RAW_FILE = RAW_DIR / "online_retail_II.csv"    
retail = pd.read_csv(RAW_FILE, encoding="ISO-8859-1")
```

```python
# =========================
# DATA OVERVIEW
# =========================

# exploring the dataframe
display(retail.head(),
retail.tail(),
retail.sample(5, random_state=42),
retail.shape,
retail.columns,
retail.info(),
retail.describe(),
retail.nunique(),
retail.info())


# exploring individual columns
print(retail["Country"].value_counts().sort_values().head(5),
retail["Description"].value_counts().sort_values().head(5))

# checking the missing values in the dataframe
print(retail.isna().sum().sort_values(ascending=False))
```
